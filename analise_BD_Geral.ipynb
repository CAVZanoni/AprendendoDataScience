```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('BD_GERAL.xlsx')
df
```

```python
df=df.iloc[:, 2:]
df
```

```python
df['Preco']=df['Preco'].str.replace('R$ ','').str.replace('.','').str.replace(',','.').astype(np.float64)
df['Tamanho']=df['Tamanho'].str.replace(' m² tot.','').astype(np.float64)
df['Preco por m2']=df['Preco']/df['Tamanho']
df
```

```python
df=df[df['Preco por m2']!=df['Preco por m2'].max()]
```

```python
plt.boxplot(df['Preco por m2'])
```

```python
# Calculando o IQR e os limites
Q1 = df['Preco por m2'].quantile(0.25)
Q3 = df['Preco por m2'].quantile(0.75)
IQR = Q3 - Q1

# Definindo os limites
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Removendo os outliers
filtro_outliers=(df['Preco por m2'] >= limite_inferior) & (df['Preco por m2'] <= limite_superior)
```
